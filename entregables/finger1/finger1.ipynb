```python
# magic function para hacer que los graficos de matplotlib se renderizen en el notebook.
%matplotlib inline 

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

Carga de un data frame

```python
props = pd.read_csv('Propiedades-0701.csv', low_memory=False)

props[:5]
```

```python
props.columns.values
```

```python
price_props_per_place = props[['place_name','price_usd_per_m2']][props['state_name'] == 'Capital Federal']
price_props_per_place
```

Habria que sacar un promedio por barrio

```python
total_props_per_place = price_props_per_place['place_name'].value_counts()
```

```python
total_price_per_place =  price_props_per_place.groupby('place_name').aggregate(sum)
total_price_per_place
```

```python
props_price_rate = pd.DataFrame({'price_usd_rate' : price_props_per_place.groupby(['place_name'])['price_usd_per_m2'].mean()}).reset_index()
ten_props_more_valuable = props_price_rate.sort_values(by='price_usd_rate',ascending=False)[:10]
```

```python
ten_props_more_valuable.plot(x='place_name',kind='bar',title='Top 10 de precios de propiedades en CABA')
plt.xlabel('Barrio Caba')
plt.ylabel('Precio promedio en U$S')
```
